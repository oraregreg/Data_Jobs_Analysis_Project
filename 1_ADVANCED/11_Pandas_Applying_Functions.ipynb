```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#to get projected salaries, with an inflation rate of 3%
#create a function

df_salary = df[pd.notna(df['salary_year_avg'])].copy()

def projected_salary(salary):
    return salary*1.03

df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
#use apply method to convert job_skills column from a list to a string
#using ast method, abstract syntax trees
import ast
ast.literal_eval(df['job_skills'][1])
```

```python
#this can also be done for the entire column

import ast

def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)  # Use skill_list parameter, not df['job_skills'][1]
    return skill_list  # Return the original NaN value

df['job_skills'] = df['job_skills'].apply(clean_list)
df['job_skills']
```

```python
#Assuming the senior roles get a 5% increase, where as the junior roles get a 3% increase in salary
#Meaning we would be affecting rows instead of columns

def projected_salary(row):
    if "Senior" in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

df_salary['salary_year_inflated'] = df_salary.apply(projected_salary, axis=1)

df_salary['salary_year_percentage'] = (df_salary['salary_year_inflated']/ df_salary['salary_year_avg'] - 1)*100
df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated', 'salary_year_percentage']].head(30)
```
