```python
#import the libraries and load the data

import pandas as pd, matplotlib.pyplot as plt
from datasets import load_dataset, Dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()  # Changed: 'to pandas' to 'to_pandas()'

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#use the sample methos, to return a random sample of items from an axis of object
df.sample(3)
```

```python
#when you sample and use the random state method, the same result will be returned every single time
#for the particular random state
df.sample(3, random_state=5)
```

```python
#dataframe copy, makes a copy of the object's indices and data

df_original = df = dataset['train'].to_pandas()
df_original
```

```python
#make a copy so that the variables are referencing the same dataframe

df_altered = df_original.copy()
df_altered.head()
```

```python
#proof that they are independent of each other
id(df_original)==id(df_altered)
```
