```python
#import the libraries and load the data

import pandas as pd, matplotlib.pyplot as plt
from datasets import load_dataset, Dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()  # Changed: 'to pandas' to 'to_pandas()'

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Fill NaN values with the median values

median_salary_year = df['salary_year_avg'].median()
```

```python
median_salary_hour = df['salary_hour_avg'].median()
```

```python
#create a new data frame with the filled values

df_filled = df
```

```python
#fill None values for salary_hour_avg
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
```

```python
df_filled.loc[:, 'salary_year_avg': 'salary_hour_avg']
```

```python
#drop duplicate removes duplicate rows
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

print('Length of original df:  ', len(df_filled))
print('Length of unique df:  ', len(df_unique))
print('Number of rows removed:  ', len(df_filled) - len(df_unique))
```

```python
#drop duplicates based off specific columns, use the subset method

df_unique = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])
print('Length of original df:  ', len(df_filled))
print('Length of unique df:  ', len(df_unique))
print('Number of rows removed:  ', len(df_filled) - len(df_unique))
```
